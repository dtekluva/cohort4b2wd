#### SCRAPE WIKI COVID TABLE

```python
covid_url = "https://en.wikipedia.org/wiki/COVID-19_pandemic_by_country_and_territory"
```

```python
response = requests.get(covid_url)
```

```python
html_content = response.content
covid_soup = bs4.BeautifulSoup(html_content)
covid_soup
```

```python
covid_container = covid_soup.find('div', {'id':'covid19-container'}) # fetch main container that caries table in question
```

```python
covid_table_body = covid_container.find('tbody')
```

```python
each_row = covid_table_body.find_all('tr')

for row in each_row:
    name_element = row.find_all('th')
    name = name_element[1].text
    
    data_element = row.find_all('td')
    if len(data_element)> 2: # CHECK IF DATA ELEMENT CONTAINS AT LEAST 3 ELEMENTS
        cases = data_element[0]
        deaths = data_element[1]
        recoveries = data_element[2]
        print(name, cases.text, deaths.text, recoveries.text)
```

#### MIGRATE WIKI COVID TABLE TO DATABASE USING PY MYSQL

```python
import pymysql.cursors

connection = pymysql.connect(user = "root",
                             password = "",
                             database = "covid",
                             host = "localhost",
                             cursorclass = pymysql.cursors.DictCursor
                            )
```

```python
with connection:
    with connection.cursor() as cursor:
        # Create a new record
        
        for row in each_row:
            name_element = row.find_all('th')
            name = name_element[1].text.replace("'", " ")

            data_element = row.find_all('td')
            if len(data_element)> 2: # CHECK IF DATA ELEMENT CONTAINS AT LEAST 3 ELEMENTS
                cases = data_element[0].text.replace(',',"").replace('\n',"") # REPLACE ALL COMAS WITH NOTHIN AND ALSO REPLACE ALL NEWLINE CHARACTERS WITH NOTHING
                deaths = data_element[1].text.replace(',',"").replace('\n',"")
                recoveries = data_element[2].text.replace(',',"").replace('\n',"")
                
                if recoveries == 'No data':
                    recoveries = 0
                if deaths == 'No data':
                    deaths = 0
                if cases == 'No data':
                    cases = 0
                print(name, cases, deaths, recoveries)
                
#                 print(name, cases.text.replace(',',"").replace('\n',""), 
#                               deaths.text.replace(',',"").replace('\n',""), 
#                               recoveries.text.replace(',',"").replace('\n',""))
                
                sql = f"INSERT INTO `stats` (`country`, `cases`, `deaths`, `recoveries`) VALUES ('{name}', '{cases}', '{deaths}', '{recoveries}')"
                cursor.execute(sql)

    # connection is not autocommit by default. So you must commit to save
    # your changes.
                connection.commit()
```
